# Aggregations

### Step 1. Import the necessary libraries

```python
import pandas as pd
import csv
```

### Step 2. Import the dataset occupation.csv from the folder or [here](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
pd.read_csv('occupation.csv')
```

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('occupation.csv',delimiter='|')
users
```

### Step 4. Discover what is the mean age per occupation

```python
u = users.groupby('occupation')
display(u[['age']].mean())

# Above can be done as below too...
# print(users.groupby('occupation').agg({'age':'mean'}))
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
import numpy as np
pd.set_option('display.max_rows', None)
m = users['gender'].eq('M')
m

users['maleRatio'] = np.where(m, m.groupby(users['occupation']).transform('mean').mul(100), 0)
users.sort_values('maleRatio',ascending = False).reset_index(drop=True)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation')['age'].agg([pd.np.min,pd.np.max])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
display(users.groupby(['occupation','gender'])['age'].agg([pd.np.mean]))

# Above can be done as below too..
display(users.groupby(['occupation','gender'])['age'].mean())
```

### Step 8.  For each occupation present the percentage of women and men

```python
# Count to gender
genderOccupation = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
genderOccupation

# Applying count for each occupation
occupationCount = users.groupby(['occupation']).agg('count')
occupationCount

# divide the genderOccupation per the occupationCount and multiply per 100
finalDF = genderOccupation.div(occupationCount, level = "occupation") * 100
finalDF

# present all rows from the 'gender column'
finalDF.loc[: , 'gender']
```
