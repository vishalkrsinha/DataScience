```python
#Trying to see, how good is our word2vec implementation

from gensim.models import word2vec,KeyedVectors, Word2Vec
import pandas as pd
import numpy as np
import nltk
```

```python
df = pd.read_csv("reddit_worldnews_sample.csv")
df.head()
```

```python
newsTitles = df['title'].values
newsTitles
```

```python
newsVec = [nltk.word_tokenize(title) for title in newsTitles] #word_tokenize breaks the text/sentence in words
newsVec
```

```python
 #Now, let's train our model to word2vec
model = Word2Vec(newsVec,min_count=1, size=32, window=5) #It's giving weight matrix stored in var - model
# size: The number of dimensions of the embeddings and the default is 100.
# window: The maximum distance between a target word and words around the target word. The default window is 5.
# min_count: The minimum count of words to consider when training the model; words with occurrence less than this count will be ignored. The default for min_count is 5.
# workers: The number of partitions during training and the default workers is 3.
# sg: The training algorithm, either CBOW(0) or skip gram(1). The default training algorithm is CBOW.
```

```python
model.wv['said']
```

```python
model.wv['rocket']
```

```python
model.wv.most_similar('man') #Close words with cosine similarity

#But, result doesn't look like a good model. reason: We didn't use sufficient size & variety of data
```

```python
vec = model['king'] - model['man'] + model['women']
model.wv.most_similar([vec])
```
