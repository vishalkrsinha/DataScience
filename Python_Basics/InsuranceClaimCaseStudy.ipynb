```python
import pandas as pd

claimsDF = pd.read_csv('claims.csv')
display(claimsDF)

customersDF = pd.read_csv('cust_demographics.csv')
display(customersDF)

customersClaimsDF = customersDF.merge(claimsDF, how = 'outer', validate='many_to_many',left_on='CUST_ID', right_on='customer_id')
display(customersClaimsDF.reset_index(drop=True))
```

```python
# 2. Perform a data audit for the datatypes and find out if there are any mismatch within the current datatypes 
# of the columns and their business significance.

customersClaimsDF.dtypes

#Audit: All float64 data type columns should have been of int data type here except 'claim_amount' column.
```

```python
# 3. Convert the column claim_amount to numeric. Use the appropriate modules/attributes to remove the $ sign.
customersClaimsDF.claim_amount = pd.to_numeric(customersClaimsDF.claim_amount.str.replace('$',''))

display(customersClaimsDF.claim_amount)
```

```python
#4. Of all the injury claims, some of them have gone unreported with the police. Create an alert flag (1,0) for all such claims.
import numpy as np

flag = customersClaimsDF.police_report

customersClaimsDF.police_report = np.where(flag == 'Yes',1,0)
customersClaimsDF.loc[customersClaimsDF.claim_type == 'Injury only'].groupby('CUST_ID')['police_report'].head()
```

```python
# 16. Is there any similarity in the amount claimed by males and females?


import scipy.stats as stats
# Independent t-test
males_spend = customersClaimsDF.claim_amount.loc[customersClaimsDF.gender == 'Male']
display(customersClaimsDF.claim_amount)

females_spend = customersClaimsDF.claim_amount.loc[customersClaimsDF.gender == 'Female']
display(females_spend)

print("The average spent by the male customers is ${}".format(males_spend.mean()))
print("The average spent by the female customers is ${}".format(females_spend.mean()))

uneq_var = stats.ttest_ind(a = males_spend, b= females_spend, equal_var=False,nan_policy='omit')  #Different variance
display(uneq_var.statistic)
display(uneq_var.pvalue)

eq_var = stats.ttest_ind(a = males_spend, b= females_spend, equal_var=True,nan_policy='omit')  #Equal variance
display(eq_var.statistic)
display(eq_var.pvalue)

# Detailed Write up: Applying Independent sample t-test as, we check influence of binary category variable over continuous variable.....
# Since the p value doesn't support the evidence of rejecting H0, will conclude that male spend == female spend. 
# Yes, there is a similarity between male and female spend.
```

```python
# 17. Is there any relationship between age category and segment?

## Ch-Square: To figure out relationship between two categorical variables...
age_seg_xtab = pd.crosstab(customersClaimsDF.Age,customersClaimsDF.Segment,margins=True)
display(age_seg_xtab)

stats.chi2_contingency(observed=age_seg_xtab)

#Detailed Write up: Applying Ch-Square as need to figure out relationship between 2 categorical variables.
#p-value is just closet to alpha, which is an evidence of failing to reject the null hypothesis. Observed == Expected, 
#meaning there is no any relationship between age category and segment!!!
```

```python
#18. The current year has shown a significant rise in claim amounts as compared to 2016-17 fiscal average which was $10,000.

display(customersClaimsDF.claim_amount.loc[pd.to_datetime(customersClaimsDF.claim_date).dt.year == 2017].mean())
display(customersClaimsDF.claim_amount.loc[pd.to_datetime(customersClaimsDF.claim_date).dt.year == 2018].mean())

ttest_1 = stats.ttest_1samp(a=customersClaimsDF.claim_amount.loc[pd.to_datetime(customersClaimsDF.claim_date).dt.year == 2018].mean(), popmean=10000)
t_score = round(ttest_1.statistic,2)
p_value = ttest_1.pvalue

print("The p value is {} and the the T Score is {}".format(p_value,t_score))

# Detailed Write up:  Applying 'One Sample T-Test' as need to figure out relationship between last year & current year data.
```

```python
# 19. Is there any difference between age groups and insurance claims?

## Chi-Square: To figure out relationship between two categorical variables...
age_claim_xtab = pd.crosstab(customersClaimsDF.Age,customersClaimsDF.claim_type,margins=True)
display(age_claim_xtab)

stats.chi2_contingency(observed=age_claim_xtab)

# Detailed Write up:  p-value is just closet to alpha, which is an evidence of failing to reject the null hypothesis. Observed == Expected, 
#meaning there is no any relationship between age category and claims!!!
```
