   ### Step 3. Assign each to a to a variable called cars1 and cars2

```python
cars1 = pd.read_csv("Exercise Data Files/Pandas Exercise Simple Joins files/cars1.csv")
display(cars1)

cars2 = pd.read_csv("Exercise Data Files/Pandas Exercise Simple Joins files/cars2.csv")
display(cars2)
```

### Step 4. It seems our first dataset has some unnamed blank columns, fix cars1

```python
cars1 = cars1.loc[:, ~cars1.columns.str.contains('^Unnamed')]
display(cars1)
```

### Step 5. What is the number of observations in each dataset?

```python
display(cars1.shape[0]) #Cars1
display(cars2.shape[0]) #Cars2
```

### Step 6. Join cars1 and cars2 into a single DataFrame called cars

```python
# pd.merge(left=cars1,
#         right=cars2,
#         left_on=['car'],
#         right_on=['car'],
#         how='right')


cars = pd.concat([cars1,cars2],ignore_index=True)
cars
```

### Step 7. There is a column missing, called owners. Create a random number Series from 15,000 to 73,000.

```python
ownersColumnSeries = np.random.randint(15000,73000,cars.shape[0])
ownersColumnSeries
```

### Step 8. Add the column owners to cars

```python
cars['owners'] = ownersColumnSeries
```
